```python
import time
import matplotlib.pyplot as plt

renderIds = [26, 41, 65, 100, 84, 169, 50, 180, 90, 278, 339, 354]


for i in renderIds:
  print(i)

  bounds = shapefile['geometry'].at[i].bounds

  maxx = bounds[2] + 0.01
  minx = bounds[0] - 0.01
  maxy = bounds[3] + 0.01
  miny = bounds[1] - 0.01

  bounding_box_2 = ee.Geometry.Rectangle(
      minx, miny, maxx, maxy
  );
  radar_tile = ee.ImageCollection('COPERNICUS/S1_GRD')\
        .filterBounds(bounding_box_2)\
        .filterDate('2022-08-01', '2022-08-31')\
        .sort('CLOUDY_PIXEL_PERCENTAGE')\
        .first()

  tile = ee.ImageCollection('COPERNICUS/S2')\
        .filterBounds(bounding_box_2)\
        .filterDate('2022-08-01', '2022-08-31')\
        .sort('CLOUDY_PIXEL_PERCENTAGE')\
        .first()


  multiband_tile = tile.select(['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B8A', 'B9', 'B10', 'B11', 'B12'])
  radar_tile = radar_tile.select(['VV', 'VH'])

  task_multiband_clouds = ee.batch.Export.image.toDrive(**{
  'image': multiband_tile,
  'description': str(i) + '_clouds',
  'folder':'ee_images/0527/august',
  'region': bounding_box_2.getInfo()['coordinates'],
  'scale': 10,
  'fileFormat':'GeoTIFF',
  'crs':'EPSG:4326',
  })

  task_radar = ee.batch.Export.image.toDrive(**{
  'image': radar_tile,
  'description': str(i) + '_sar',
  'folder':'ee_images/0527/august',
  'region': bounding_box_2.getInfo()['coordinates'],
  'scale': 10,
  'fileFormat':'GeoTIFF',
  'crs':'EPSG:4326',
  })

  # starting tasks
  task_multiband_clouds.start()
  while task_multiband_clouds.active():
    print('Polling for task active - clouds')
    time.sleep(3)

  task_radar.start()
  while task_radar.active():
    print('Polling for task active - radar')
    time.sleep(3)

  print('Polling done')

```
